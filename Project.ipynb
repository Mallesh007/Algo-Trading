# Checking SMA for these stocks

```python
for i in stock_close:
    stock_close[i+"_SMA50"]=stock_close[i].rolling(window=50,min_periods=50).mean()
    stock_close[i+"_SMA200"]=stock_close[i].rolling(window=200,min_periods=200).mean()
    stock_close[i+"_EMA75"]=stock_close[i].ewm(span=75,min_periods=75).mean()
    stock_close[i+"_EMA100"]=stock_close[i].ewm(span=100,min_periods=100).mean()
    stock_close[i+"_EMA150"]=stock_close[i].ewm(span=150,min_periods=150).mean()
```

# Cummulative Returns, Drawdowns 

```python
ticker=["AAPL","JNJ","UNH","LMT","NFLX"] 
```

```python
stocks=yf.download(ticker)
```

```python
stocks
```

```python
stock_close=stocks.Close
```

```python
stock_close
```

```python
stock_close.dropna(inplace=True)
```

```python
stock_close
```

```python
for i in stocks.Close:
    stock_close[i+"d_returns"]=np.log(stock_close[i].div(stock_close[i].shift(1)))
```

```python
stock_close.dropna(inplace=True)
```

```python
stock_close
```

```python
for i in stocks.Close:
    print(i)
    print(stock_close[i+"d_returns"].sum())
    print(np.exp(stock_close[i+"d_returns"].sum()))
```

```python
for i in stocks.Close:
    stock_close[i+"_cummreturns"]=stock_close[i+"d_returns"].cumsum().apply(np.exp)
```

```python
stock_close
```

```python
for i in stocks.Close:
    stock_close[i+"_cummreturns"].plot(figsize=(12,8),title=i+" buy and hold",fontsize=12)
    plt.legend(loc="upper left")
    plt.show()
```

```python
for i in stocks.Close:
    stock_close[i+"_cummax"]=stock_close[i+"_cummreturns"].cummax()
```

```python
stock_close
```

```python
for i in stocks.Close:
    stock_close.loc[:,[(i+"_cummreturns"),(i+"_cummax")]].plot(figsize=(12,8),title=i+" buy and hold",fontsize=12)
    plt.legend(loc="upper left")
    plt.show()
```

```python
for i in stocks.Close:
    stock_close[i+"_drawdown"]=stock_close[i+"_cummax"]-stock_close[i+"_cummreturns"]
```

```python
stock_close
```

```python
for i in stocks.Close:
    print(i)
    print(stock_close[i+"_drawdown"].max())
```

```python
for i in stocks.Close:
    print(i)
    print(stock_close[i+"_drawdown"].idxmax())
```

```python
for i in stocks.Close:
    stock_close[i+"_drawdown%"]=(stock_close[i+"_cummax"]-stock_close[i+"_cummreturns"])/stock_close[i+"_cummax"]
```

```python
stock_close
```

```python
for i in stocks.Close:
    print(i)
    print(stock_close[i+"_drawdown%"].max())
```

```python
for i in stocks.Close:
    print(i)
    print(stock_close[i+"_drawdown%"].idxmax())
```
